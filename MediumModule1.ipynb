```python
import pandas as pd
import plotly.express as px
import folium

df = pd.read_csv('/content/drive/MyDrive/Large Data Sets for Python/worlddata.csv')


wdiselect = df[(df['IndicatorName']=="Population, ages 0-14 (% of total)") \
                       & (df['Year']==1960)]
wdiselect
```

```python
data_to_plot = wdiselect[['CountryCode','Value']]
data_to_plot.head()
```

```python
from urllib.request import urlopen
import json
counties = json.load(urlopen('https://raw.githubusercontent.com/plotly/datasets/master/geojson-counties-fips.json'))

import pandas as pd
df2 = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/fips-unemp-16.csv", dtype={"fips": str})


fig = px.choropleth(data_to_plot, locations='CountryCode', color='Value',
                           color_continuous_scale="Viridis",
                           range_color=(data_to_plot.Value.min(), data_to_plot.Value.max()),
                           labels={'Population, ages 0-14 (% of total)'}
                          )
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

```python
wdiselect2 = df[(df['IndicatorName']=="Population, ages 0-14 (% of total)") \
                       & (df['Year']==2014)]

data_to_plot2 = wdiselect2[['CountryCode','Value']]

fig2 = px.choropleth(data_to_plot2, locations='CountryCode', color='Value',
                           color_continuous_scale="Viridis",
                           range_color=(data_to_plot2.Value.min(), data_to_plot2.Value.max()),
                           labels={'Population, ages 0-14 (% of total)'}
                          )
fig2.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig2.show()
```

```python
wdiselect3 = df[(df['IndicatorName']=="Population, ages 0-14 (% of total)") |
                    (df['IndicatorName']=="Adolescent fertility rate (births per 1,000 women ages 15-19)") |
                    (df['IndicatorName']=="Population, total")]


wdiselect3 = wdiselect3.dropna()

pivot_df = wdiselect3.pivot_table(index=['CountryCode', 'Year'], columns='IndicatorName', values='Value').reset_index()
pivot_df = pivot_df.dropna()


fig3 = px.scatter(pivot_df, x="Population, ages 0-14 (% of total)",
                     y="Adolescent fertility rate (births per 1,000 women ages 15-19)",
                     animation_frame="Year",
                     animation_group="CountryCode",
                     size="Population, total",
                     hover_name="CountryCode")


fig3.show()
```
